```python
# 第3章: 正規表現
import urllib.request
urllib.request.urlretrieve('http://www.cl.ecei.tohoku.ac.jp/nlp100/data/jawiki-country.json.gz', 'jawiki-country.json.gz')

import gzip
with gzip.open('jawiki-country.json.gz', mode='rb') as gziped_f, open ('jawiki-country.json', mode='wb') as json_f:
    json_f.write(gziped_f.read())
```

```python
# 20. JSONデータの読み込み
import json

article_dicts = []
with open('jawiki-country.json', mode='rb') as json_f:
    for json_line in json_f:
        article_dicts.append(json.loads(json_line))
    
for article_dict in article_dicts:
    if article_dict['title'] == 'イギリス':
        print(article_dict['text'])
```

```python
# 21. カテゴリ名を含む行を抽出
import re

category_prog = re.compile('(\[\[Category:)(.*)(\|?.*)(\]\])')

for article_dict in article_dicts:
    article_text = article_dict['text']
    article_lines = article_text.splitlines()
    for article_line in article_lines:
        category_match = category_prog.fullmatch(article_line)
        if category_match:
            print(article_line)
```

```python
# 22. カテゴリ名の抽出
category_prog2 = re.compile('(\[\[Category:)([^|]+)(\|?)(.*)(\]\])')

for article_dict in article_dicts:
    article_text = article_dict['text']
    article_lines = article_text.splitlines()
    for article_line in article_lines:
        category_match = category_prog2.fullmatch(article_line)
        if category_match:
            print(category_match.groups()[1])
```

```python
# 23. セクション構造
section_prog = re.compile(r'(=+)([^=]+)(=+)')

for article_dict in article_dicts:
    article_text = article_dict['text']
    article_lines = article_text.splitlines()
    for article_line in article_lines:
        section_match = section_prog.fullmatch(article_line)
        if section_match:
            section_level = len(section_match.groups()[0]) - 1
            section_name = section_match.groups()[1]
            print('%s, %s' % (section_level, section_name))  
```

```python
# 24. ファイル参照の抽出
reffile_prog = re.compile('(\[\[ファイル:)(.+)(\|.+)*(\]\])')

for article_dict in article_dicts:
    article_text = article_dict['text']
    article_lines = article_text.splitlines()
    for article_line in article_lines:
        reffile_match = reffile_prog.match(article_line)
        if reffile_match:
            print(reffile_match.groups()[1])
```
